```python
import csv
tww = {}
with open('census.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        if row["STNAME"] in tww:
          tww[row["STNAME"]].append((row["CTYNAME"], row["CENSUS2010POP"]))
        else:
            tww[row["STNAME"]] = []
```

```python
for x in tww:
  print("No of cities in", x, "is", len(tww[x]))
```

```python
highest_city =  max(tww, key=lambda x: len(tww[x]))
lowest_city = min(tww, key=lambda x: len(tww[x]))
print("The state that has the most cities is", highest_city, "with", len(tww[highest_city]), "cities.")
print("The state that has the least cities is", lowest_city, "with", len(tww[lowest_city]), "cities.")
```

```python
for x in tww:
  print("The total population of", x, "is", sum(int(y[1]) for y in tww[x]))
  
```

```python
for x  in tww:
  max_city = max(tww[x], key=lambda x: int(x[1]))
  print("The city with the highest population in", x, "is", max_city[0], "with a population of", max_city[1])
  
  lowest_city = min(tww[x], key=lambda x: int(x[1]))
  print("The city with the lowest population in", x, "is", lowest_city[0], "with a population of", lowest_city[1],end ="\n\n")
  
```

```python
with open('US2010Cencus.csv', mode='w', newline='') as csv_file:
  csv_writer = csv.writer(csv_file)
  csv_writer.writerow(['State', 'Total City', 'Total Population'])
  for x in tww:
    csv_writer.writerow([x, len(tww[x]), sum(int(y[1]) for y in tww[x])])
```
